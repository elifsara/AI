```python
import pandas as pd
```

```python
df=pd.read_csv('ExampleData.csv')
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
imp=IterativeImputer(estimator=lr)
imputed_data=imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data)
```

```python
from sklearn.impute import KNNImputer #En yakın komşu ile doldurma
```

```python
knn=KNNImputer()
```

```python
imputed_data2=knn.fit_transform(df)
```

```python
pd.DataFrame(imputed_data2, columns=[df.columns])
```

```python
from sklearn.impute import SimpleImputer #Orta değerle doldurma
```

```python
imp=SimpleImputer(strategy='mean')  #median, mode, ...(e çok tekrar eden değerle orta değerle doldurma)
```

```python
imputed_data3=imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data3,columns=[df.columns])
```
