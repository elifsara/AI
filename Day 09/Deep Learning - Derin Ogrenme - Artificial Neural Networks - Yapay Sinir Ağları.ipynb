```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x=df.drop('Outcome', axis=1)
y=df[['Outcome']]     
#3 farklı yol ile yapılabilir
```

```python
df.columns
```

```python
x=df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
y=df[['Outcome']]
```

```python
x=df.iloc[:,0:8]
y=df.iloc[:,8]
```

```python
x.head()
```

```python
y.head()
```

```python
model=Sequential()
model.add(Dense(120,activation='relu'))
model.add(Dense(80,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1,activation='sigmoid'))  #sigmoid classification olduğunu söylüyor
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) #hata fonksiyonu - loss fonction
#adam - sonuca doğru yaklaştığında adımları küçültüyor.
```

```python
model.fit(x,y,epochs=250, batch_size=32, validation_split=.20, verbose=1)
#epoch- kaç defa gidip geldiğini söylüyor. 
#validation_split verinin ne kadarının test için ayrıldığını gösteriyor.
#batch_size= tek seferde kaç satır taşıyacağını gösterir.
```

```python
import pandas as pd
```

```python
df=pd.read_pickle('kc_house.pkl')  #feature engineering yapılmış hali
```

```python
df.head()  
```

```python
x=df.drop('price', axis=1)
y=df[['price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=.20,random_state=42)
```

```python
model=Sequential()
model.add(Dense(120,activation='relu'))
model.add(Dense(80,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(4,activation='relu'))
model.add(Dense(1))  #regression
model.compile(loss='mse', optimizer='adam',) 
```

```python
history=model.fit(x_train,y_train, validation_data=(x_test,y_test), batch_size=128, epochs=1500, verbose=0)
 # başına %timeit yazınca kaç sn sürdüğünü gösteriyor. Ama time peketini indirmen gerek
# sonuna , verbose=0 yazdığında rakamlar görünmüyor. 
```

```python
tahmin=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(y_test,tahmin)
```

```python
mean_squared_error(y_test,tahmin)**.5
```

```python
loss_f=pd.DataFrame(history.history)
```

```python
loss_f.plot()
```
