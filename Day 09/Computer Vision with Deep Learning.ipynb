# Computer Vision with Deep Learning

```python
import tensorflow as tf
```

```python
mnist=tf.keras.datasets.mnist  #el yazması rakamlari tanıma
```

```python
(train_images,train_labels),(test_images,test_labels)=mnist.load_data()
```

```python
#normalize data    
#resimlerde herbir piksel 0-255 arası değer aldığı için 
#her değeri 255'e bölünce değerler 0 -1 arasaında olacağınddan normalize etmiş oluruz
```

```python
train_images=train_images/255.0
test_images=test_images/255.0
```

```python
#Convolutional Neural Networks - CNNs Image Classification Resim sınıflandırma modeli
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, InputLayer, Reshape, MaxPooling2D, Flatten, Dropout, BatchNormalization
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(28,28)))
model.add(Reshape(target_shape=(28,28,1)))
model.add(Conv2D(filters=12,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(10)) # 10 fakli cevap classification 0-9 a kadar olan rakamlar
model.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, validation_split=.10, epochs=10)
```

```python
loss, accuracy=model.evaluate(test_images, test_labels)
```

```python
accuracy
```

```python
model.save('benimelyazmasitaniyanmodelim.h5')
```
