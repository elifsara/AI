```python
import spacy
from spacy import displacy
NER=spacy.load('en_core_web_sm')
```

```python
text="""ELif Sara loves NLP and artificial intelligence. Python is a great programming language.
The Turkish Space Agency is the national space agency of Turke, headquartered in Ankara. 
It operates under the Department of Space which is directly overseen
by the President of Turkey and administor of TUA is Serdar Yıldırım."""
```

```python
text1=NER(text)
```

```python
for word in text1.ents:
    print(word.text,word.label_)
```

```python
spacy.explain('LOC')
```

```python
displacy.render(text1)
```

```python
displacy.render(text1,style='ent', jupyter=True)
```
